```python
#10.a. Implementation of word embedding - Bag of Words

from sklearn.feature_extraction.text import CountVectorizer
corpus=["I love machine learning",
        "Machine learning is fun",
        "I love programming"]

vectorizer=CountVectorizer()
X=vectorizer.fit_transform(corpus)

bow_matrix=X.toarray()
vocabulary=vectorizer.get_feature_names_out()
print("Bag of Words Matrix:")
print(bow_matrix)
print("\nVocabulary:")
print(vocabulary)
```

```python
#10.b. Implementation of word embedding - TF-IDF

import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

corpus = [
    'I love programming in Python',
    'Python programming is fun',
    'I love coding in Python',

]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
tfidf_array = X.toarray()
feature_names = vectorizer.get_feature_names_out()

print("TF-IDF Representation:")
print(tfidf_array)

print("\nFeature Names (Words):")
print(feature_names)


for i, doc in enumerate(tfidf_array):
    print(f"\nDocument {i + 1}:")
    for j, score in enumerate(doc):
        if score > 0:
            print(f"{feature_names[j]}: {score:.4f}")
```
